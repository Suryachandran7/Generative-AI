```python
df1["total_sqft"]=df1["total_sqft"].astype(str)
```

```python
def sqft(x):
    y=x.split("-")
    if len(y)==2:
        return(float(y[0])+float(y[1]))/2
    elif len(y)==1:
        return x
    else:
        return None
```

```python
df1["total_sqft"]=df1["total_sqft"].apply(sqft)
```
