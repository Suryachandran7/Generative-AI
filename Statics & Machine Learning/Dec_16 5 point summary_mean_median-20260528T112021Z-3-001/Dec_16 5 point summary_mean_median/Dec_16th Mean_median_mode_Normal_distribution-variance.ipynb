# Calculating Sample Variance in Python




```python
from statistics import variance 

#define data
data = [4, 8, 12, 15, 9, 6, 14, 18, 12, 9, 16, 17, 17, 20, 14]

#calculate sample variance
variance(data)
```

```python
#calculating Population Variance in Python
```

```python
from statistics import pvariance 

#define data
data = [4, 8, 12, 15, 9, 6, 14, 18, 12, 9, 16, 17, 17, 20, 14]

#calculate sample variance
pvariance(data)
```

```python
print(np.var (data))
```

```python
#numpy gives the population variance
```

```python
from statistics import stdev
```

```python
# sample standard deviation
```

```python
stdev(data)
```

```python
# population standard deviation
```

```python
from statistics import pstdev
```

```python
pstdev(data)
```

# IQR

```python
num_arr=np.random.randint(0,25,500)
```

```python
num_arr
```

# IQR

```python
# calculating 25 percentile
Q1=np.percentile(num_arr,25)
Q1
```

```python
# when we added the value with 5 , percentile also added with 5
Q1=np.percentile(num_arr+5,25)
Q1
```

```python
# when we increase value by 5% percentile also increase by 5%
Q1=np.percentile(1.05*(num_arr),25)
Q1
```

```python
# when we increase value by 50 %,  percentile also increase by 50%
Q1=np.percentile(1.5*(num_arr),25)
Q1
```

```python
# calculating 50 percentile
Q2=np.percentile(num_arr,50)
Q2
```

```python
Q2=np.percentile(1.05*(num_arr),50)
Q2
```

```python
# calculating 75 percentile
Q3=np.percentile(num_arr,75)
Q3
```

```python

Q3=np.percentile(1.05*(num_arr),75)
Q3
```

```python
# percentile and quantile work on sorted list
```

```python
# percentile and quantile indicate the place in the list .
```

```python
IQR=Q3-Q1
```

```python
IQR
```
