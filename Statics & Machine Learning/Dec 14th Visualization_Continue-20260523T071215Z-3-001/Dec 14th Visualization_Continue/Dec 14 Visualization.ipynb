```python
#countplot plot the count of the number of records by category
```

```python
#gender distribution
plt.figure(figsize=(10,6))
sns.countplot(x=empdf["Gender"])
plt.title("Distribution of Gender in the company")
plt.show()
```

```python
# as per observation there are approx 600 no female & 900 no of male employees.
```

```python
#business travel distribution
plt.figure(figsize=(10,6))
sns.countplot(x=empdf["BusinessTravel"])
plt.title("Distribution of BusinessTravel in the company")
plt.show()
```

```python
#department distribution
plt.figure(figsize=(10,6))
sns.countplot(x=empdf["Department"],hue=empdf["Gender"])
plt.title("Distribution of Department in the company")
plt.show()
```

```python
#work life balance distribution
plt.figure(figsize=(10,6))
sns.countplot(x=empdf["WorkLifeBalance"])
plt.title("Distribution of work life balance in the company")
plt.show()
```

```python
#work life balance distribution
plt.figure(figsize=(10,6))
sns.countplot(x=empdf["MonthlyRate"])
plt.title("Distribution of work life balance in the company")
plt.show()
```

```python
#gender distribution as per Attrition
plt.figure(figsize=(10,6))
sns.countplot(x=empdf["Gender"],hue=empdf["Attrition"])
plt.title("Attriction level based on gender")
plt.show()
```

```python
#gender distribution as per Attrition
plt.figure(figsize=(10,6))
sns.countplot(x=empdf["Gender"],hue=empdf["MaritalStatus"],palette=sns.color_palette("RdBu"))
plt.title("MaritalStatus level based on gender")
plt.show()
```

```python
#gender distribution as per Attrition
plt.figure(figsize=(10,6))
sns.countplot(x=empdf["MaritalStatus"],hue=empdf["Attrition"])
plt.title("Attriction level based on MartialStatus")
plt.show()
```

```python
plt.subplots(1,3,figsize=(20,10))

plt.subplot(1,3,1)

sns.countplot(x=empdf["MaritalStatus"],hue=empdf["Attrition"])
plt.title("Attriction level based on MartialStatus")

plt.subplot(1,3,2)

sns.countplot(x=empdf["Gender"],hue=empdf["MaritalStatus"],palette=sns.color_palette("RdBu"))
plt.title("MaritalStatus level based on gender")

plt.subplot(1,3,3)

sns.countplot(x=empdf["Gender"],hue=empdf["MaritalStatus"],palette=sns.color_palette("RdBu"))
plt.title("MaritalStatus level based on gender")




plt.show()
```

# BAR PLOT

```python
# bar plot or bar chart represents how a numerical variable relates to a categorical variables
```

```python
plt.figure(figsize=(14,6))
ax=sns.barplot(data=empdf,x="EducationField",y="MonthlyRate")
```

```python
plt.figure(figsize=(14,6))

ax=sns.barplot(x=empdf["EducationField"],y=empdf["MonthlyRate"])
ax.set(ylabel="MONTHLYRATE")
plt.ylabel("MONTHLYRATE",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(14,6))

ax=sns.barplot(x=empdf["EducationField"],y=empdf["MonthlyRate"],hue="Gender",data=empdf)
ax.set(ylabel="MONTHLYRATE")
plt.ylabel("MONTHLYRATE",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(14,6))

ax=sns.barplot(x=empdf["EducationField"],y=empdf["MonthlyRate"],hue="JobRole",data=empdf)
ax.set(ylabel="MONTHLYRATE")
plt.ylabel("MONTHLYRATE",fontsize=20)
plt.show()
```
