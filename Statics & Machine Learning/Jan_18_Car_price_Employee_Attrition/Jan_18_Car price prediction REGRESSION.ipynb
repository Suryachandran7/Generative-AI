```python
"""
From the above display we can see that:
1)In the 'Mileage' column there is 'km' written, we need to seperate this 'km' from mileage.
2)In the 'Doors' column there are month (-May) names which we need to remove.
3)The 'Levy' column contains '-' symbol. We need to remove it from this column.
"""
```

```python
#(1)Handling Mileage column
```

```python
#removing the 'km' from the mileage column and converting it to float
```

```python
df["Mileage"]=df["Mileage"].str.replace("km","")
```

```python
df["Mileage"]=df["Mileage"].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
#replacing all the '0' values with the mean values of the 'Mileage' column
df['Mileage'] = np.where(df['Mileage'] == 0.0,df['Mileage'].mean(),df['Mileage'])
```

```python
df.describe()
```

```python
#(2)Handling the Doors column 
```

```python
#checking the unique values of 'Doors' column
df['Doors'].unique()
```

```python
df["Doors"]=df["Doors"].str.replace("-May","")
```

```python
df["Doors"]=df["Doors"].str.replace("-Mar","")
```

```python
df.info()
```

```python
df['Doors'].unique()
```

```python
df['Doors']=df['Doors'].map({"02":2,"04":4,">5":5})
```

```python
#(3)Handling Levy column
```

```python
#checking the unique values of 'Levy' column after cleaning
df['Levy'].unique()
```

```python
df['Levy'] = pd.to_numeric(df['Levy'].replace('-', '0'))
```

```python
#Replacing the 0 in the 'Levy' column with mean of that column
df['Levy'] = np.where(df['Levy'] == 0.0,df['Levy'].mean(),df['Levy'])
```

```python
#handling Engine volume column
```

```python
#checking the unique values in the 'Engine volume' column
df['Engine volume'].unique()
```

```python
#We can see that there are some values with 'Turbo' and some values without 'Turbo'
#So we remove the word 'Turbo' from all records that have it

df['Engine volume'] = pd.to_numeric(df['Engine volume'].str.split(' ').str[0], downcast='float')
```

```python
#Replacing the '0' in the 'Engine volume' column with the mean value of that column
df['Engine volume']=np.where(df['Engine volume'] == 0.0,df['Engine volume'].mean(),df['Engine volume'])
```

```python
#Feature engineering the production year column
import datetime as dt
currt_time = dt.datetime.now()
df['year_used'] = currt_time.year - df['Prod. year'] 
```

```python
import statsmodels.api as sm
X10_final = sm.add_constant(X10_scaled)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
variable=X10_final
VIF=pd.DataFrame()
VIF["Variance Inflation Fcator"]=[variance_inflation_factor(variable.values,i)for i in range(variable.shape[1])]
```
