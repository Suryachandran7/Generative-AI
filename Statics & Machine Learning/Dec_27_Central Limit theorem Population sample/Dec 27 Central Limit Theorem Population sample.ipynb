# Central limit theorem

```python
"""
The Central Limit Theorem (CLT) is a statistical theory that indicate
that the mean and standard deviation derived from a sample, 
will accurately approximate the mean and standard deviation of the population.
The minimum number of members in the sample is 30 or more than 30.
is 30
"""
```

# law of large number

```python
"""# law of large number
#(1)sample size = or more than 30.
#(2)Higher the sample size , lower the error .
If the sample size is large enough more than 30 , the sample 
distribution taken from any population distribution regradless of 
its shape will approximately a normal distribution.
# law of large number.

The sample size increases the meqan of the sample is close to the mean of the population.
no. of element in the sample >=30

"""
```

```python
# creating samples of size 200 and histogram of samples 
```

```python
#Create a list
sampled_means200 = []
# For 100 times:
for i in range(len(cardio2)):
    # Take a random sample of 200 rows from the popoulation, take the mean of these rows,append to sampled_means
    sampled_means200.append(cardio2.sample(200).mean())
# plotting histogram
plt.figure(dpi = 140) #resolution of the figure
sns.histplot(sampled_means200,kde=True)
plt.xlabel('Cholesterol')
plt.ylabel('Frequency')
plt.title("Histogram: Mean of samples of Cholesterol")
plt.show()
```

```python
# creating samples of size 100 and histogram of samples 
```

```python
#Create a list
sampled_means100 = []
# For 100 times:
for i in range(len(cardio2)):
    # Take a random sample of 100 rows from the popoulation, take the mean of these rows,append to sampled_means
    sampled_means100.append(cardio2.sample(100).mean())
# plotting histogram
plt.figure(dpi = 140) #resolution of the figure
sns.histplot(sampled_means100,kde=True)
plt.xlabel('Cholesterol')
plt.ylabel('Frequency')
plt.title("Histogram: Mean of samples of Cholesterol")
plt.show()
```

```python
# the sample of 100  histogram is well normal distribution with respect to population.
```

```python
# View the mean of the sampled_means
pd.Series(sampled_means100).mean()
```

```python
# The population mean 246 and sample mean 245.92 is very close when the sample size is 100
```

```python
# Subtract Mean Sample Mean from True population mean
error = cardio2.mean()- pd.Series(sampled_means100).mean()
#print
print('The Mean Sample Mean is only %f different the True Population(cardio2) mean!' % error)
```
